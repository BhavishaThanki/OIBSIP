### <i> Dataset </i>

```python
path = "C:\Program Files\Datasets\CarPrice.csv"
df=pd.read_csv(path)
df
```

```python
df.dtypes
```

### <i>Dropping unnecessary/extra features </i>

```python
UnCols = ["car_ID","symboling","doornumber","drivewheel","enginelocation","cylindernumber","boreratio","stroke","compressionratio"]

df2= df.drop(UnCols,axis=1)
df2
```

### <i>One Hot Encoding for CarName attribute

```python
df3=pd.get_dummies(df2,"CarName")
df3
```

### <i> Removing target column

```python
data = df3.drop("price",axis=1)
y= df3["price"]
```

### <i> Splitting data and fitting model

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(data, y, test_size=0.2 , random_state=42)

from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(xtrain, ytrain)
```

###  <i> Predictions on test data

```python
predictions = model.predict(xtest)
predictions
```

### <i> Actual values

```python
ytest.to_numpy()
```

### <i> Accuracy score

```python
model.score(xtest,ytest)
```
